Find and print the indices of pairs of elements in the following array “a” for which the value of their sum is a multiple of 7.

a = np.array([1, 7, 3, 10, 14, 76, 4])

```python
import numpy as np

a = np.array([1, 7, 3, 10, 14, 76, 4])
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if (a[i] + a[j]) % 7 == 0:
            print((i, j))
```

Calculate the inverse of matrix A (save it as A_inv) and its dot product and element-wise multiplication with B using NumPy.

    A = np.array([[1, 2, 0],[0, 5, 6],[7, 0, 9]])
    B = np.array([[5, 6, 7],[7, 8, 9],[1, 2, 3]])


```python
import numpy as np

A = np.array([[1, 2, 0],
              [0, 5, 6],
              [7, 0, 9]])

B = np.array([[5, 6, 7],
              [7, 8, 9],
              [1, 2, 3]])

# Inverse of A
A_inverse = np.linalg.inv(A)

# Dot product
dot_product = np.dot(A, B)

# Element-wise multiplication
element_wise = A * B

print("A inverse:")
print(A_inverse)

print("\nDot Product:")
print(dot_product)

print("\nElement-wise Multiplication:")
print(element_wise)
```

Compute the eigen values the matrix A using np.linalg.eig . Find the spectral radius of the matrix A.

A = np.array([[1, 2, 0],[0, 5, 6],[7, 0, 9]])

```python
import numpy as np

A = np.array([[1, 2, 0],
              [0, 5, 6],
              [7, 0, 9]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

spectral_radius = np.max(np.abs(eigenvalues))

print("\nSpectral Radius:")
print(spectral_radius)
```

Consider the matrix
              B = np.array([[5, 6, 7], [7, 8, 9], [1, 2, 3]])  ;   Using NumPy, answer the following:

(a) Compute the SVD.
 Use np.linalg.svd(B, full_matrices=False) to obtain the matrices U, S, and Vt. Convert the one-dimensional array S into a diagonal matrix Sigma = np.diag(S).

(b) Reconstruct the original matrix as B_recon = U @ Sigma @ Vt. Test the accuracy of the reconstruction with np.allclose(B, B_recon, atol=1e-10) and report the Boolean result.

(c) Using the singular values in S, compute the spectral (2-norm) condition number
κ2B=σmaxσmin
    Print the value of κ2B


```python
import numpy as np

B = np.array([[5, 6, 7],
              [7, 8, 9],
              [1, 2, 3]])

# (a) SVD
U, S, Vt = np.linalg.svd(B, full_matrices=False)

Sigma = np.diag(S)

print("U =")
print(U)

print("\nS =")
print(S)

print("\nSigma =")
print(Sigma)

print("\nVt =")
print(Vt)

# (b) Reconstruction
B_recon = U @ Sigma @ Vt

print("\nReconstructed B =")
print(B_recon)

print("\nReconstruction Accurate?")
print(np.allclose(B, B_recon, atol=1e-10))

# (c) Condition Number
kappa_2_B = S.max() / S.min()

print("\nCondition Number (kappa_2(B)) =")
print(kappa_2_B)
```
