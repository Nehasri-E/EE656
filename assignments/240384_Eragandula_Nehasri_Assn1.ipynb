Write a Python program that performs the following tasks:
a)Accept a sentence input from the user.
b)Print the sentence in uppercase.
c)Count and print the number of vowels in the sentence.
d)Create and print a new sentence by reversing the order of words in the original sentence.

```python
#Accept a sentence input from the user.
sentence = input("Enter a sentence: ")
#Print the sentence in uppercase.
print("Uppercase sentence:", sentence.upper())
#Count and print the number of vowels in the sentence.
vowels = 'aeiouAEIOU'
count = 0
for char in sentence:
  if char in vowels:
    count +=1
print("Number of vowels:", count)
#Reversing the order of words in the original sentence.
words = sentence.split()
new_sentence = " ".join(words[ : : -1])
print("The new sentence is:", new_sentence)
```

Write a Python program that does the following:
a)Using a ‘for’ loop, print all integers starting from 1 up to a user-provided number (let's call it ‘n’) that are divisible by 3.
b)Using a ‘while’ loop, print the Fibonacci series up to a user-provided term number (let's call it ‘m’).

```python
# All integers starting from 1 up to n that are divisible by 3.
n = int(input("Enter a number:"))
print("Numbers that are divisible by 3 are: ")
for i in range(1,n+1):
  if i % 3 == 0:
    print(i)
```

```python
# The Fibonacci series up to m
m = int(input("Enter a number:"))
print("The Fibonacci series: ")
a = 0
b = 1
count = 0
while count < m:
 print(a)
 next_term = a+b
 a = b
 b = next_term

 count = count + 1
```

You are managing an inventory of a store. Given a dictionary ‘inventory’ representing the stock of items, perform the following tasks:
*  Print the current inventory.
*  Update the quantity of "Apples" to 20.
*  Add a new item "Grapes" with a quantity of 10.
*  Remove the item "Bananas" from the inventory.
*  Print the final inventory.








```python
# The current inventory
Inventory = {}

n = int(input("Enter the number of items:"))
for i in range(n):
  item = input("Enter the item name:")
  quantity = int(input("Enter the quantity:"))
  Inventory[item] = quantity
print("The current inventory is:", Inventory)

# Update quantity of Apples
Inventory["Apples"] = 20

# Adding a new item Grapes
Inventory["Grapes"] = 10

# Removing Bananas
del Inventory["Bananas"]

# The final Inventory
print("The final Inventory is:", Inventory)
```

Write a Python program that simulates a basic data filtering task using functions and lists:
*  Define a function named process_data that accepts a list of integers as its argument.
*   Inside the function, calculate and print the mean (average) of the numbers in the list.
*   Use a list comprehension to create a new list containing only the numbers from the original list that are strictly greater than the calculated mean.
*    Have the function return a tuple containing the mean and the newly filtered list and print this output outside the function.




```python
# Function definition
def process_data(numbers):
  mean = sum(numbers) / len(numbers)
  print("The mean is : " , mean)
  filtered_list = [num for num in numbers if num > mean]
  return[mean, filtered_list]

nums = list(map(int, input("Enter numbers: ").split()))

result = process_data(nums)

print("The Returned Output is: ", result)
```
